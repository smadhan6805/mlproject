```python
text_columns = ['blood pressure', 'sugar']
def convert_text_to_numeric(df, column):
    df_data[column] = pd.to_numeric(df_data[column], errors='coerce')

for column in text_columns:
    convert_text_to_numeric(df_data, column)
    print(f"{column} : {df_data[column].dtype}")
```

```python
missing_values = df_data.isnull().sum()
print(missing_values)
```

```python
def mean_value_imputation(df, column):
    mean_value = df[column].mean()
    df_data[column].fillna(value = mean_value, inplace=True)

def mode_value_imputation(df_data, column):
  mode = df_data[column].mode()[0]
  df_data[column] = df_data[column].fillna(mode)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.linear_model import LogisticRegression  # Example models
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Assuming df_data is already loaded and preprocessed
print("Column names in the DataFrame:", df_data.columns)
# Ensure the correct column name is used for the target variable
# Replace 'class' with the actual target column name if different
target_column = "classification"

# Ensure the target column exists
if target_column in df_data.columns:
    # Separate features and target
    x = df_data.drop(target_column, axis=1)
    y = df_data[target_column]

    # One-Hot Encode categorical columns (non-numeric) if necessary
    x = pd.get_dummies(x, drop_first=True)

    # Split the data into training and testing sets
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

    # Define the 'models' list with model instances and their names
    models = [
        ('Logistic Regression', LogisticRegression()),
        ('Decision Tree', DecisionTreeClassifier()),
        ('Random Forest', RandomForestClassifier())
    ]

    # Now you can proceed with model training and evaluation
    for name, model in models:
        print(f"{name} model")
        print()

        # Fit the model
        model.fit(x_train, y_train)

        # Predict the test set
        y_pred = model.predict(x_test)

        # Print confusion matrix
        print("Confusion Matrix:")
        print(confusion_matrix(y_test, y_pred))
        print('\n')

        # Print various evaluation metrics
        print('Accuracy Score:', accuracy_score(y_test, y_pred))
        print('Precision Score:', precision_score(y_test, y_pred))
        print('Recall Score:', recall_score(y_test, y_pred))
        print('F1 Score:', f1_score(y_test, y_pred))
        print('\n')
```
